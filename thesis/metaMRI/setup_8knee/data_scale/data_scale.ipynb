```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data1 = []
data2 = []

for i in range(1,6):
    # read csv
    filename1 = f"/cheng/metaMRI/metaMRI/thesis/metaMRI/setup_8knee/data_scale/log/run-metaMRI_thesis_metaMRI_setup_8knee_data_scale_E_sup_maml_1few-adapt_lr0.001_stanford_seed{i}-tag-Testing SSIM.csv"
    df1 = pd.read_csv(filename1)
    # extract value
    values1 = df1['Value'].values
    data1.append(values1)
    # read csv
    filename2 = f"/cheng/metaMRI/metaMRI/thesis/metaMRI/setup_8knee/data_scale/log/run-metaMRI_thesis_metaMRI_setup_8knee_data_scale_E_sup_standard_1few-adapt_lr0.001_stanford_seed{i}-tag-Testing SSIM.csv"
    df2 = pd.read_csv(filename2)
    # extract value
    values2 = df2['Value'].values
    data2.append(values2)


data1 = [sublist[:70] for sublist in data1]
data2 = [sublist[:70] for sublist in data2]

# mean and std
mean_values1 = np.mean(data1, axis=0)
std_values1 = np.std(data1, axis=0)
mean_values2 = np.mean(data2, axis=0)
std_values2 = np.std(data2, axis=0)
```

```python
min_index2 = np.argmin(mean_values2)
print('Standard min NMSE: ',  round(mean_values2[min_index2], 5))
print('Standard stand diviation: ', round(std_values2[min_index2], 5))
print('Standard epoch: ', min_index2)

min_index1 = np.argmin(mean_values1)
print('MAML min NMSE: ',  round(mean_values1[min_index1], 5))
print('MAML stand diviation: ', round(std_values1[min_index1], 5))
print('MAML epoch: ', min_index1)
```

```python
max_index2 = np.argmax(mean_values2)
print('Standard max SSIM: ',  round(mean_values2[max_index2], 5))
print('Standard stand diviation: ', round(std_values2[max_index2], 5))
print('Standard epoch: ', max_index2)

max_index1 = np.argmax(mean_values1)
print('MAML max SSIM: ',  round(mean_values1[max_index1], 5))
print('MAML stand diviation: ', round(std_values1[max_index1], 5))
print('MAML epoch: ', max_index1)
```
