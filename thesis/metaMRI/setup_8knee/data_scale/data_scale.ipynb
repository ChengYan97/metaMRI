```python
import matplotlib.pyplot as plt
import numpy as np

# 创建示例数据
x = np.arange(0, 10, 1)
y = np.sin(x)
variance = np.random.rand(len(x)) * 0.2  # 示例方差数据

# 绘制折线图
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Data')

# 在某个点上标注方差，并添加上下误差线
index_to_highlight = 5  # 要标注方差的点的索引
plt.errorbar(x[index_to_highlight], y[index_to_highlight], yerr=variance[index_to_highlight],
             fmt='ro', label='Point with Variance', capsize=5, capthick=2)

# 添加标签和图例
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# 显示图形
plt.show()
```

```python
standard = [0.01733, 0.01547, 0.01462,0.01392 ,0.01322, 0.01286, 0.01245]
maml = [0.016694, 0.01522, 0.01437, 0.01381, 0.01313, 0.01279, 0.01244]
x = [1, 5, 10, 20, 50, 100, 200]
variance_standard = [0, 0.0001, 0, 0, 0, 0]
variance_maml = [0.000187, 0.0001, 0, 0, 0, 0]

plt.figure(figsize=(6,5),dpi=400)

# alpha: transparency 
plt.plot(x, maml, color=colors[0], marker='o',markersize=4, label='Initialized by MAML train on P')
plt.errorbar(x[0], maml[0], yerr=variance_maml[0], fmt=colors[0], capsize=3, capthick=1)
plt.errorbar(x[1], maml[1], yerr=variance_maml[1], fmt=colors[0], capsize=3, capthick=1)

plt.plot(x, standard, color=colors[1], marker='o',markersize=4, label='Initialized by standard train on P')
plt.errorbar(x[1], standard[1], yerr=variance_standard[1], fmt=colors[1], capsize=3, capthick=1)

plt.xscale('log')  # Set x-axis to log scale
plt.xticks(x,x)
plt.tick_params(axis='both', which='major', labelsize=FONTSIZE/2)
plt.tick_params(axis='both', which='minor', labelsize=(FONTSIZE-2)/2)
plt.xlabel('Data scale for adaptation',fontsize=FONTSIZE/2)
plt.ylabel('Best NMSE on target domain',fontsize=FONTSIZE/2)
#plt.title('Out-of-distribution Q3 Adaptation', fontsize = FONTSIZE/2)
plt.legend(fontsize = (FONTSIZE-3)/2)
plt.grid(True)
plt.show()
```
