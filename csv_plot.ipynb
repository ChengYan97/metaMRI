```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data1 = []
data2 = []

for i in range(1,4):
    # read csv
    filename1 = f"/cheng/metaMRI/metaMRI/save/E6.4/run-metaMRI_save_E6.4_E6.4_maml_adapt_Q1_seed{i}-tag-Testing SSIM.csv"
    df1 = pd.read_csv(filename1)
    # extract value
    values1 = df1['Value'].values
    data1.append(values1)
    # read csv
    filename2 = f"/cheng/metaMRI/metaMRI/save/E6.4/run-metaMRI_save_E6.4_E6.4_standard_adapt_Q1_seed{i}-tag-Testing SSIM.csv"
    df2 = pd.read_csv(filename2)
    # extract value
    values2 = df2['Value'].values
    data2.append(values2)

# mean and std
mean_values1 = np.mean(data1, axis=0)
std_values1 = np.std(data1, axis=0)
mean_values2 = np.mean(data2, axis=0)
std_values2 = np.std(data2, axis=0)

# alpha: transparency 
plt.plot(range(0, 31), mean_values1, color='red', label='MAML')
plt.fill_between(range(0, 31), mean_values1 - std_values1, mean_values1 + std_values1, color='red', alpha=0.15, edgecolor=None)
plt.plot(range(0, 31), mean_values2, color='blue', label='Standard')
plt.fill_between(range(0, 31), mean_values2 - std_values2, mean_values2 + std_values2, color='blue', alpha=0.15, edgecolor=None)

plt.xlabel('Epochs')
plt.ylabel('SSIM')
plt.title('Out-of-distribution Q3 Adaptation')
plt.legend()
plt.show()
```

```python
import math
import torch
import torch.optim as optim
from torch.optim.lr_scheduler import CosineAnnealingLR
import torchvision.models as models


model = models.resnet18(pretrained=False)

optimizer = optim.SGD(model.parameters(), lr=0.001)

total_iterations = 200
initial_lr = 0.001

scheduler = CosineAnnealingLR(optimizer, total_iterations/1, eta_min=0.0001, last_epoch=-1)

lr = []
for iteration in range(total_iterations):
    
    scheduler.step()
    
    current_lr = optimizer.param_groups[0]['lr']
    lr.append(current_lr)
```

```python
#plt.figure(figsize=(6,5))
plt.figure(figsize=(6,5))
plt.plot(list(range(1, len(lr)+1)), lr, label='cosine annealing lr', linewidth=2)
plt.legend()
plt.yticks([i * 0.0001 for i in range(11)])
plt.xlabel('Epoch',fontsize=15)
plt.ylabel('Learning rate',fontsize=15)
# plt.savefig('save/plots/fasterrcnn_v2.png', format='png', bbox_inches='tight')
```
